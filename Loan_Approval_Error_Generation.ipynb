```python
import pandas as pd
import numpy as np
import random

# Load the dataset
df = pd.read_csv("C:/Users/Ananya/Downloads/sem 2/DSP/loan_approval_data.csv")

# 1. Random String Injection in annual_income column
def introduce_string_into_income(value):
    if random.random() < 0.02:  # 2% chance
        return random.choice(["N/A", "Unknown"])
    return value

# 2. Negative or Unrealistic Value in loan_amount column
def introduce_negative_or_high_loan(value):
    if random.random() < 0.02:
        return random.choice([-1 * random.randint(100000, 1000000), random.randint(100000000, 500000000)])
    return value

# 3. Incorrect Category in education column
def introduce_wrong_education(value):
    if random.random() < 0.02:
        return random.choice(["PhD", "Undergrad", "Vocational"])
    return value

# 4. Random Date in Numerical Field for loan_term
def introduce_date_in_loan_term(value):
    if random.random() < 0.02:
        return random.choice(["01-01-2023", "12-31-2025", "15-07-2021"])
    return value

# 5. Unrealistic Values in cibil_score
def introduce_unrealistic_cibil(value):
    if random.random() < 0.02:
        return random.choice([-50, 1200, 1500])
    return value

# 6. Missing Values in dependents
def introduce_missing_dependents(value):
    if random.random() < 0.02:
        return np.nan
    return value

# 7. Switching Values in employment and loan_status
def introduce_switch_employment(value):
    if random.random() < 0.02:
        return "Yes" if value == "No" else "No"
    return value

def introduce_switch_loan_status(value):
    if random.random() < 0.02:
        return "Approved" if value == "Rejected" else "Rejected"
    return value

# 8. Numerical Category Injection in loan_status
def introduce_numeric_in_loan_status(value):
    if random.random() < 0.02:
        return random.choice([0, 1])  # Representing loan status as numbers
    return value
# Remove leading and trailing spaces from column names
df.columns = df.columns.str.strip()

# Verify column names again to ensure no spaces remain
print("Cleaned column names:\n", df.columns)

# Now, re-run the error-introduction code with cleaned column names
df['annual_income'] = df['annual_income'].apply(introduce_string_into_income)
df['loan_amount'] = df['loan_amount'].apply(introduce_negative_or_high_loan)
df['education'] = df['education'].apply(introduce_wrong_education)
df['loan_term'] = df['loan_term'].apply(introduce_date_in_loan_term)
df['cibil_score'] = df['cibil_score'].apply(introduce_unrealistic_cibil)
df['dependents'] = df['dependents'].apply(introduce_missing_dependents)
df['employment'] = df['employment'].apply(introduce_switch_employment)
df['loan_status'] = df['loan_status'].apply(introduce_switch_loan_status).apply(introduce_numeric_in_loan_status)


# Save the modified DataFrame with errors to a new CSV file
df.to_csv("C:/Users/Ananya/Downloads/sem 2/DSP/loan_approval_data_with_errors.csv", index=False)
```
